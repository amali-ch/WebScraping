**Web Scraping**

four basic steps:
--> sending a http get request to the url of the webpage by using request library

--> fetching and parsing the data usinf beautifulsoup and maintain the data in some dict or list

--> analyzing the HTML tags and their attributes

-->output the data in any file format(.csv,.xlsx,.json)

```python
import requests
from bs4 import BeautifulSoup
```

```python
#enter the url from makaan.com
url= "https://www.makaan.com/hyderabad-residential-property/buy-property-in-hyderabad-city"
```

```python
request=requests.get(url)
request
```

```python
request.content
```

```python
soup=BeautifulSoup(request.content,'html')
#print(soup)
print(type(soup))
```

```python
print(soup.prettify())#it also shows content but in a prettier way
```

```python
sel_lst1=soup.find_all('span',attrs={'class':'seller-type'})
#i ll append all the sellertype into a list
a1=[]
for i in sel_lst1:
  a1.append(i.text)
print(a1
      )
```

```python
sel_lst2=soup.find_all('div',attrs={'data-type':'price-link'})
#i ll append all the sellertype into a list
a2=[]
for i in sel_lst2:
  a2.append(i.text)
print(a2)
```

```python
sel_lst3=soup.find_all('span',attrs={'class':'cityName'})
#i ll append all the sellertype into a list
a3=[]
for i in sel_lst3:
  a3.append(i.text)
print(a3)
```

```python
sel_lst4=soup.find_all('td',attrs={'class':'lbl rate'})
#i ll append all the sellertype into a list
a4=[i.text for i in sel_lst4]
print(a4)
```

```python
sel_lst5=soup.find_all('div',attrs={'data-type':'price-link'})
#i ll append all the sellertype into a list
a5=[i.text for i in sel_lst5]
print(a5)
```

```python
sel_lst6=soup.find_all('div',attrs={'class':'title-line'})
#i ll append all the sellertype into a list
a6=[i.span.text.strip() for i in sel_lst6]
print(a6)
```

```python
#no of bathrooms is not mentioned....... so to get no of brs
bathr=soup.find_all('li',attrs={'class':'keypoint','title':'Bathrooms'})
bath_info=[i.text for i in bathr]
print(bath_info)
```

```python
#i ll mention NaN if not mentioned
bathr=soup.find_all('ul',attrs={'class':'listing-details'})
bath_info=[]
for i in bathr:
  if i.find('li',attrs={'class':'keypoint','title':'Bathrooms'}):
    bath_info.append(i.text)
  else:
    bath_info.append("NaN")
print(bath_info)      
```

```python
import pandas as pd
p=pd.DataFrame({
    'seller':a1,'price':a2,'city':a3,'lbl rate':a4,'bhk':a5,'bathR':bath_info
    
})
print(p)
```

```python
p.to_csv('webscrap.csv')
```
